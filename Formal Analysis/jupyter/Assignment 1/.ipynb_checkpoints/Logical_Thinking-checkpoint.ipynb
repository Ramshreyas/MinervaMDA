### Would you rather donate your body to science or donate your organs to people who need them?

**PART 1**

For one side of the argument, craft two brief arguments (<150 words each) using deductive reasoning. One argument should be a valid deduction and one should contain a fallacy. Write these arguments in natural language in simple terms with clear use of logical connectives and atomic sentences so that it is amenable for the following analysis.

> I would rather donate my body to science than donate my organs to people who need them.

Valid deduction:

If donating to science helps more people than donating directly, I'd rather donate to science  
The organs in my body are more than or equal to the number of people I can help by donating directly  
Donating to science will help more people than there are organs in my body  
If donating to science will help more people than there are organs in my body, and the organs in my body are more than or equal to the people I can help by donating directly, donating to science will help more people than donating directly  
I'd rather donate my organs to science  

1. Explain why these are deductive arguments by identifying the features of deductive reasoning present (<75 words).  

The argument is composed of premises and a conclusion.  
The premises are all declarative sentences - they are either true or false.  
The conclusion follows from the premises.

2. For the valid deduction:  
a. Translate your argument into symbolic logic.

A:  Donating to science will help more people than there are organs in my body

B:  Donating to science will help more people than donating directly  
DS: I'd rather donate my organs to science  

B -> DS  
O  
A  
A & O -> B  
DS
